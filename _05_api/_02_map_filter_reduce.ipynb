## map
iterable객체의 각요소에 대해 지정한 함수를 적용하고, 그 결과를 map객체로 반환
```
map(function, iterable) -> map
```




````python
nums: list = [1,2,3,4,5]
result = map(lambda x: x ** 2, nums)

print(result, type(result)) # map 주소, map 타입
print("list(result): ", list(result))

# map() : 기존 list의 요소를 하나씩 꺼내 함수를 호출,
#         함수 결과들을 모아 새로운 list 생성
````

````python
# 문자열 리스트 -> 문자열 길이 리스트
````

````python
products: list[str] = ['아이스 아메리카노', '수박주스', '말차라떼', '식혜', '딸기스무디']

products_len: list[int] = map(lambda x: len(x), products)
print(products_len)
print(list(products_len))
````

````python
# 여러 순회 가능 객체(list) 통합
mid_term = [90, 55, 70]
final_term = [100, 40, 75]

avg_scores = list(map(lambda x, y: (x + y) / 2, mid_term, final_term))
print(avg_scores)
````

## filter

주어진 iterable객체에 대해서 참/거짓 평가후 참인 요소들만 결과값으로 반환

```
filter(function, iterable) -> filter
```

````python
nums = list(range(1, 11))

evens = list(filter(lambda n: n % 2 == 0, nums))
oods = list(filter(lambda n: n % 2 == 1, nums)) # list(filter(lambda n: n % 2 != 0, nums))
threes = list(filter(lambda n: n % 3 == 0, nums))

print(evens)
print(oods)
print(threes)
````

````python
# 이름 목록에서 성이 "홍" 씨인 사람만 찾아서 list로 반환
names: list[str] = ['홍길동', '고길동', '홍진경', '최홍만','유재석', '홍명보']

# str.startwith(str2) -> bool
# - 문자열의 시작이 str2와 같아면 True
hong_family = list(filter(lambda name: name.startswith('홍') , names))
print(hong_family)
````

````python
# dict 검색(dict 필터링)
# 학생 목록에서 점수가 90점 이상인 학생만 필터링
students = [
    { 'name' : '이순신', 'score' : 80},
    { 'name' : '윤봉길', 'score' : 75},
    { 'name' : '논개', 'score' : 99},
    { 'name' : '강감찬', 'score' : 91},
]

# filter() 함수
# students1 = list(filter(lambda score: score >= 90, students))
# TypeError: '>=' not supported between instances of 'dict' and 'int'

students1 = list(filter(lambda s: s['score'] >= 90, students))
print(students1)

#list comprehension
students2 = [s for s in students if s['score'] >= 90]
print(students2)
````

## reduce
iterable객체의 요소에 대한 연산결과를 단일값을 반환.
합계, 최대/최소값 등의 연산에 적합

```
reduce(function, iterable, initial_value) -> Any
```
함수는 3개의 인자를 받는다.
1. 누적 연산값
2. 현재 요소
3. 초기값 (선택)


````python
from functools import reduce

# 정수형 list 내 요소들의 합계
nums: list[int] = [11,22,33,44,55,66,77,88,99]

# 총합 저장
# total = reduce(콜백함수(누적용변수, 현재값), 순회객체, 누적변수초기값)
# -> 콜백 함수의 return 값이 자동으로 누적용 변수에 대입 된다
total = reduce(lambda sum, curr: sum + curr, nums, 0)
print("total: ", total)
````

````python
# 최대, 최소 값 찾기
nums: list[int] = [23,64,4735,4629,6734,694,5743,-3,-3745,-2345,-8473]

max_num: int = reduce(lambda max, curr: max if max > curr else curr, nums, 0)
print(max_num)

min_num: int = reduce(lambda min, curr: min if min < curr else curr, nums, 0)
print(min_num)

# 절대값이 가장 큰 수 + 일반 함수
def abs_func(max, curr):
    print(f'max: {max}, curr: {curr}')
    return max if abs(max) > abs(curr) else curr

abs_max = reduce(abs_func, nums, 0)
print("abs_max: ", abs_max)
````

````python
# 최고 점수를 지닌 학생 출력

students = [
    { 'name' : '이순신', 'score' : 80},
    { 'name' : '윤봉길', 'score' : 75},
    { 'name' : '논개', 'score' : 99},
    { 'name' : '강감찬', 'score' : 91},
]

# high_score = reduce(lambda s, curr: s['score'] if s['score'] > curr['score'] else curr['score'], students)
# TypeError: 'int' object is not subscriptable

# 초기 값 생략 -> 첫 번째 값 대입
func = lambda s, curr: s if s['score'] > curr['score'] else curr
high_score = reduce(func, students)

print("결과 : ", high_score)
# 결과 : {'name' : '논개', 'score' : 99}
````
