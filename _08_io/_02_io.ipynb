```python
# - with절
# - with 옆에 작성된 open() 구문에서 사용한 파일 연결을
#   with절 종료 시 자동으로 close

with open('helloworld2.txt', 'w', encoding='utf-8') as f:
    f.write('with을 이용하면 자동 close()된다')
# print(f.read())

with open('helloworld.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

## 파일 읽기

```python
# 한줄씩 읽기
with open('helloworld.txt', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        print(f'i : {i} / line : {line}', end='')  # end='' => print완료 후 마지막 문자 빈칸
```

```python
# 파일 이어쓰기('a', appen)
with open('helloworld.txt', 'a', encoding='utf-8') as f:
    f.write('''ㅋㅋㅋ
ㅎㅎㅎㅎ
ㅌㅌㅌㅌ
ㅊㅊㅊㅊ''')

with open('helloworld.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# 파일복사
# 1. 파일일기
with open('helloworld.txt', 'r', encoding='utf-8') as f:
    text = f.read()
# 2. 파일쓰기
with open('helloworld_copy.txt', 'w', encoding='utf-8') as f:
    f.write(text)
    print("파일 복사 완료")
```

```python
# 이미지 파일 복사(아무 파일이나 가능)
# 문자는 : r,w,a는 문자열만 가능
# 문자가 아닌 파일들은: rb,wb 바이트단위
with open('image_test.png', 'rb') as img:
    imb_bites = img.read()
    print(imb_bites)
with open('image_test_copy.png', 'wb') as img:
    img.write(imb_bites)
    print('이미지 복사 완료')
```

## dict-csv 읽고 쓰기

```python
# csv 읽기
# csv.dictreader(f)
# csv.DictWriter(f, fieldnames)
import csv

with open('users.csv', 'r', encoding='utf-8') as _reader:
    reader = csv.DictReader(_reader)  # DictReader 객체 생성
    # for row in reader:
    #     print(row, type(row))
    #     users.append(row)

    users = [row for row in reader]
    # users[0].items()
    print(users)
```

```python
# csv 쓰기
with open('users.csv', 'w', encoding='utf-8', newline='') as _writer:
    # - fieldnames: 첫 줄에 작성될 헤더(컬럼명)
    writer = csv.DictWriter(_writer, fieldnames=['id', 'name', 'email'])

    # 필드 네임 헤더등록
    writer.writeheader()

    #users에 사용자 추가
    users.append({'id': 400, 'name': '이순신', 'email': 'lee@gmail.com'})
    users.append({'id': 500, 'name': '감강찬', 'email': 'kam@gmail.com'})

    # 여러줄 쓰기
    # writer.writerows(users)
    # 한 줄씩 쓰기
    for user in users:
        writer.writerow(user)

    print(users)

```
