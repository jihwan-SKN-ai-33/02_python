# while 반복문

- `while` 반복문은 조건식이 `True`로 평가되는 동안 코드를 반복 실행한다.
- 조건식이 `False`가 되면 반복을 중지한다.
- 반복 횟수가 미리 정해져 있지 않고, 특정 조건이 만족될 때까지 반복해야 할 때 자주 사용한다.

## 1. 기본 while 구조

`while`은 보통 초기값, 조건식, 증감식을 함께 사용한다.

```python
i = 1
while i <= 3:
    print(i)
    i += 1
```

## 2. break로 반복 중지하기

`while True`는 조건식이 항상 참인 반복문이다. 이 구조를 사용할 때는 반드시 반복을 빠져나올 수 있는 `break` 조건을 만들어야 한다.
- break (분기문) : 가까운 반복문 종료
ㄷ

````python
# 1 ~ 10 까지 출력 (무한루프 + bradk)
i = 1
while True: # while 문은 True일 때
    print(i)
    i += 1

    if i == 11:
        break
````

````python
# continue : 반복 순서를 즉시 다음으로 넘어감

for x in range(3):
    print("앞", x)

    if x == 1:
        print("if", x)
        print("continue 등장!")
        continue

    print("뒤", x)
````

## 3. 메뉴 주문 예제

````python
menu = """
===========
구내식당 메뉴
===========
1. 제육볶음
2. 김밥
3. 김치찌개
4. 된장찌개
0. 종료(break)
===========
"""

print(menu)

orders: list[str] = []
while True:
    sel = int(input("메뉴 선택 : "))
    print("선택: ", sel)

    match sel:
        case 1:
            orders.append("제육볶음")
        case 2:
            orders.append("김밥")
        case 3:
            orders.append("김치찌개")
        case 4:
            orders.append("된장찌개")
        case 0:
            orders.append("메뉴 선택 종료")
            break # while 종료
        case _:
            print("메뉴에 작성된 번호만 입력하세요.")

print(f'주문 내역: {orders}')
````
