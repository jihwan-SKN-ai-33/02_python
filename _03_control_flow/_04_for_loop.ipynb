# for 반복문

- `for` 반복문은 iterable 객체의 요소를 하나씩 꺼내 실행한다.
- `str`, `list`, `tuple`, `dict`, `set`, `range` 등이 대표적인 iterable 객체이다.
- 반복 횟수가 정해져 있거나 컬렉션의 모든 요소를 순회할 때 자주 사용한다.

## 1. 문자열, 리스트, 튜플 순회

문자열은 문자 단위로, 리스트와 튜플은 요소 단위로 순회한다. (시퀀스 타입)

```python
text: str = 'hello world'
for ch in text:
    print(ch)

lst: list[int] = [10, 20, 30]
for num in lst:
    print(num)

tpl: tuple[int] = (30, 40, 50)
for num in tpl:
    print(num)
```

## 2. dict와 set 순회

`dict`를 그대로 순회하면 key가 나온다. `items()`를 사용하면 key와 value를 함께 꺼낼 수 있다.

`set`은 순서가 없는 자료형이므로 출력 순서는 실행 환경에 따라 달라질 수 있다.

```python
from typing import Any

# type hint - dict[str, Any]에서 Any
# - 어떤 타입이든 작성 가능하다는 것을 나타내는 표현(class)
user_info: dict[str, Any] = {
    'name': '홍길동',
    'age': 23,
    'hobby': ['독서', '유뷰트']
}

# dict 순회 -> key 순회, value 순회, item 순회(+unapcking)
print('--- dick key 순회 ---')
# for A in B
# A : 객체에서 요소 하나만 꺼내와 저장할 변수
# B : 순회가능한 객체 (srt, list, tuple, dice, set, ...)

for key in user_info:
    print(f'{key}: {user_info[key]}')

print('--- dict item 순회 ---')
for key, value in user_info.items():
    print(f'{key}: {value}')

print('--- set 순회 : if를 곁들인 ---')
nums: set[int] = {4,8,1,2,4,33,4,1}

# 결과 저장용 list
results: list[int]  = []

for num in nums:
    if num % 2 == 0:
        results.append(num) # 결과 저장용 list에 추가

print(results)


```

## 3. range 객체

`range(start, end, step)`은 정수 범위를 표현하는 객체이다.

- `start`: 시작값
- `end`: 끝값, 포함되지 않음
- `step`: 증가 또는 감소 간격

```python
# range 객체 :
# 지정된 범위 내 정수를 step 만큼 증가하면서 값을 하나씩 반환하는 객체


print("--- range 객체 ---")
rng = range(1, 5, 1) # 1 이상 5 미만
print(rng, type(rng))

for num in rng:
    print(num)

print('--- range 객체를 이용한 정수 list 만들기 ---')
nums: list[int] = list(range(1, 11, 1))
print("nums: ", nums)

print('--- 1부터 100까지 모든 정수 합 구하기 ---')

# n의 합계를 누적할 변수
n_sum = 0

for n in range(1, 101, 1):
    n_sum += n

print("n_sum: ", n_sum)

print("--- 1부터 100까지 홀수의 합 구하기 ---")
n_sum = 0

for n in range(1, 100, 2):
    n_sum += n
print("n_sum: ", n_sum)

n_sum = 0
for n in range(1, 101, 1):
    if n % 2 == 1:
        n_sum += n
print("n_sum: ", n_sum)
```

## 4. 중첩 반복문

반복문 안에 반복문을 넣으면 행과 열처럼 2차원 구조를 만들 수 있다. 구구단과 별찍기 같은 예제에 자주 사용한다.

```python
print("--- 중첩 반복문 예시 ---")
for a in range(1,4):
    for b in range(1,3):
        print(a, b)
```
