### **문제 10**

다음 조건에 맞는 프로그램을 작성하세요.

1. 두 셋 `A = {1, 2, 3, 4}`와 `B = {3, 4, 5, 6}`가 있습니다.
2. `A`와 `B`의 **대칭 차집합**을 구하세요.
3. 결과를 `print()`를 사용해 출력하세요

````python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

AB = A.symmetric_difference(B)

print("대칭 차집합: ",AB )
````
